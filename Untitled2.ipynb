```python
import winsound
import time

# Define the melody notes (frequency in Hz) and durations (in ms)
song = [
    # "Take me out to the ball game"
    (262, 400), (262, 400), (440, 400), (392, 400), (330, 400), (392, 800),  # C C A G E G
    (294, 800),                                                             # D

    # "Take me out with the crowd"
    (262, 400), (262, 400), (440, 400), (392, 400), (330, 400), (392, 400),  # C C A G E G
    (392, 800),                                                             # G

    # "Buy me some peanuts and cracker jack"
    (440, 400), (466, 400), (440, 400), (330, 400), (349, 400), (392, 400),  # A G# A E F G
    (440, 800),                                                             # A

    # "I don't care if I never get back"
    (440, 400), (440, 400), (494, 400), (523, 400), (494, 400), (440, 400),  # A A B C B A
    (392, 800),                                                             # G

    # "Root, root, root for the home team"
    (262, 400), (262, 400), (440, 400), (392, 400), (330, 400), (392, 800),  # C C A G E G
    (294, 800),                                                             # D

    # "If they don't win, it's a shame"
    (440, 400), (440, 400), (392, 400), (349, 400), (330, 800),             # A A G F E
    (294, 400),                                                             # D

    # "For it's one, two, three strikes you're out"
    (440, 400), (440, 400), (440, 400), (392, 400), (349, 400), (330, 400),  # A A A G F E
    (294, 800),                                                             # D

    # "At the old ball game!"
    (262, 400), (262, 400), (440, 400), (392, 400), (330, 400), (392, 800),  # C C A G E G
    (294, 800)                                                              # D
]

# Function to play the song
def play_song(song):
    for freq, dur in song:
        if freq == 0:  # Rest
            time.sleep(dur / 1000.0)
        else:
            winsound.Beep(freq, dur)

# Play the song
play_song(song)
```
