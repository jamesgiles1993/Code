```python
import pulp

# create the LP object, set up as a maximization problem
lp = pulp.LpProblem('Home Run Maximization', pulp.LpMaximize)

# create the player variables
players = {
    'player1': {'prior': 20, 'projected': 30},
    'player2': {'prior': 30, 'projected': 40},
    'player3': {'prior': 15, 'projected': 25},
    'player4': {'prior': 10, 'projected': 20},
    'player5': {'prior': 25, 'projected': 35},
    'player6': {'prior': 18, 'projected': 28},
    'player7': {'prior': 20, 'projected': 30},
    'player8': {'prior': 20, 'projected': 30},
    'player9': {'prior': 22, 'projected': 32},
    'player10': {'prior': 27, 'projected': 37},
    'player11': {'prior': 12, 'projected': 22},
    'player12': {'prior': 15, 'projected': 25},
    'player13': {'prior': 17, 'projected': 27},
    'player14': {'prior': 13, 'projected': 23},
    'player15': {'prior': 23, 'projected': 33},
    'player16': {'prior': 21, 'projected': 31},
    'player17': {'prior': 16, 'projected': 26},
    'player18': {'prior': 11, 'projected': 21},
    'player19': {'prior': 28, 'projected': 38},
    'player20': {'prior': 19, 'projected': 29}
}
x = pulp.LpVariable.dicts('x', players.keys(), lowBound=0, cat='Binary')

# create the constraint for the sum of prior year's home runs
lp += sum(players[player]['prior'] * x[player] for player in players) <= 168

# create the constraint for the number of players in the lineup
lp += sum(x[player] for player in players) == 8

# create the objective function to maximize
lp += sum(players[player]['projected'] * x[player] for player in players)

# solve the LP
lp.solve()

# print the solution
print('Projected number of home runs for each player in the lineup:')
for player in players:
    if x[player].varValue == 1:
        print(player, ':', players[player]['projected'])

print('Total projected number of home runs:', pulp.value(lp.objective))
```

```python
from pulp import *

def optimize_lineup(players, n=1):
    # Create a binary variable for each player indicating whether they are in the lineup or not
    player_vars = LpVariable.dicts("Players", players, lowBound=0, upBound=1, cat='Integer')

    # Create the 'prob' variable to contain the problem data
    prob = LpProblem("FantasyBaseballLineupOptimization", LpMaximize)

    # Add the objective function to maximize points
    prob += lpSum([players[player]["points"] * player_vars[player] for player in players])

    # Add constraints to enforce the position requirements
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "C"]) == 1
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "1B"]) == 1
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "2B"]) == 1
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "3B"]) == 1
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "SS"]) == 1
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "OF"]) == 3
    prob += lpSum([player_vars[player] for player in players if players[player]["position"] == "P"]) == 2

    # Add constraint to enforce the salary cap
    prob += lpSum([players[player]["salary"] * player_vars[player] for player in players]) <= 50000



    # Solve the optimization problem
    prob.solve()

    # Get the best n unique lineups
    lineups = []
    for i in range(n):
        lineup = [player for player in players if value(player_vars[player]) == 1]
        if lineup in lineups:
            continue
        lineups.append(lineup)
        print(f"Lineup {i+1}:")
        print(f"{'Player':<10} {'Points':<10} {'Position':<10} {'Team':<10}")
        for player in lineup:
            print(f"{player:<10} {players[player]['points']:<10} {players[player]['position']:<10} {players[player]['team']:<10}")
        print()

        # Count the number of players on each team in the lineup
        team_counts = {}
        for player in lineup:
            if players[player]["team"] not in team_counts:
                team_counts[players[player]["team"]] = 1
            else:
                team_counts[players[player]["team"]] += 1
               
        print(team_counts)
        # Add constraint to enforce having exactly three unique teams in a lineup
        prob += lpSum([team_counts[team] == 5 for team in team_counts]) == 1
        
        # Add a constraint to exclude this lineup from future solutions
        prob += lpSum([player_vars[player] for player in lineup]) <= len(lineup) - 1
        
        prob.solve()

    # Print the status of the optimization problem
    print(f"Status: {LpStatus[prob.status]}")
```
