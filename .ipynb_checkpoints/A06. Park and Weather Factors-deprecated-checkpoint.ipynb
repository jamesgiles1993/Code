```python
average_dictionary = {
    'b1': 0.143953, 
    'b2': 0.044871,
    'b3': 0.004155,
    'hr': 0.032105,
    'bb': 0.083775,
    'hbp': 0.010398,
    'so': 0.219759,
    'fo': 0.121683,
    'go': 0.233031,
    'lo': 0.058949,
    'po': 0.047321
    }
```

```python
def create_park_and_weather_factors(game_df, date, overwrite_year=None):
     # Extract daily games
    daily_game_df = game_df.query(f'date == "{date}"').reset_index(drop=True)

    daily_weather_list = []
    # Loop over games
    for i in range(len(daily_game_df)):
        # Extract relevant information
        game_id = daily_game_df['game_id'][i]
        venue_id = daily_game_df['venue_id'][i]
        away_team = daily_game_df['away_team'][i]
        home_team = daily_game_df['home_team'][i]
        datetime = daily_game_df['game_datetime'][i]
        date = daily_game_df['date'][i]
        year = daily_game_df['year'][i]
        
        # Extract weather from box score
        weather, wind, park, full_date = create_box(game_id)

        
        # Create game weather dataframe
        game_weather_dictionary = {
            'game_id': game_id,
            'away_team': away_team,
            'home_team': home_team,
            'venue_id': venue_id,
            'park': park,
            'datetime': datetime,
            'date': date,
            'year': year,
            'weather': weather,
            'wind': wind
        }
        game_weather_df = pd.DataFrame(game_weather_dictionary, index=[0])
        game_weather_df = clean_weather(game_weather_df)

        daily_weather_list.append(game_weather_df)

    # Create day weather dataframe
    daily_weather_df = pd.concat(daily_weather_list, axis=0).reset_index(drop=True)

    # Overwrite year (might be useful early in the season):
    if overwrite_year is not None:
        daily_weather_df['year'] = overwrite_year
    
    # Columns to keep
    keep_columns = list(daily_weather_df.columns)
    
    # Add park dummies
    active_venues = list(team_map['VENUE_ID'])
    active_venues = [str(venue) for venue in active_venues]
    for park in active_venues:
        daily_weather_df[f'venue_{park}'] = (daily_weather_df['venue_id'].astype(str) == park).astype(int)
    active_venue_columns = [f"venue_{park}" for park in active_venues]

    for year in range(2015, 2025):
        daily_weather_df[f'year_{year}'] = (daily_weather_df['year'].astype(int) == year).astype(int)
    year_inputs = [f'year_{year}' for year in range(2015, 2025)]


    # Loop over lefty/righty dummy
    for lefty_dummy in [0,1]:
        # Assign lefty dummy
        daily_weather_df['lefty'] = lefty_dummy

        if lefty_dummy == 0:
            side = 'r'
        else:
            side = 'l'

        # Step 2: Create interaction terms
        interaction_terms = []
        # Weather x Park
        for col1 in ['x_vect', 'y_vect', 'temperature']:
            for col2 in active_venue_columns:
                interaction_name = col1 + '_' + col2
                daily_weather_df[interaction_name] = daily_weather_df[col1] * daily_weather_df[col2]
                interaction_terms.append(interaction_name)
        # Weather x Park x batSide
        for col1 in ['x_vect', 'y_vect', 'temperature']:
            for col2 in active_venue_columns:
                for col3 in ['lefty']:
                    interaction_name = col1 + '_' + col2 + '_' + col3
                    daily_weather_df[interaction_name] = daily_weather_df[col1] * daily_weather_df[col2] * daily_weather_df[col3]
                    interaction_terms.append(interaction_name)
        # Year x Park
        for col1 in year_inputs:
            for col2 in active_venue_columns:
                interaction_name = col1 + '_' + col2
                daily_weather_df[interaction_name] = daily_weather_df[col1] * daily_weather_df[col2]
                interaction_terms.append(interaction_name)
        # Weather x Park x batSide
        for col1 in year_inputs:
            for col2 in active_venue_columns:
                for col3 in ['lefty']:
                    interaction_name = col1 + '_' + col2 + '_' + col3
                    daily_weather_df[interaction_name] = daily_weather_df[col1] * daily_weather_df[col2] * daily_weather_df[col3]
                    interaction_terms.append(interaction_name)
        # batSide x Park
        for col1 in ['lefty']:
            for col2 in active_venue_columns:
                interaction_name = col1 + '_' + col2
                daily_weather_df[interaction_name] = daily_weather_df[col1] * daily_weather_df[col2]
                interaction_terms.append(interaction_name)
        daily_weather_df['x_vect_temperature'] = np.abs(daily_weather_df['x_vect']) * daily_weather_df['temperature']
        daily_weather_df['y_vect_temperature'] = daily_weather_df['y_vect'] * daily_weather_df['temperature']
        daily_weather_df['x_vect_y_vect'] = np.abs(daily_weather_df['x_vect']) * daily_weather_df['y_vect']
        interaction_terms.extend(['x_vect_temperature', 'y_vect_temperature', 'x_vect_y_vect'])
    
    
        # Add averages
        for event in events_list:
            daily_weather_df[f'{event}_b'] = average_dictionary[event]
            daily_weather_df[f'{event}_p'] = average_dictionary[event]
            
            # Step 6: Calculate deciles based on predicted        
            model = globals().get(event + '_model')
            daily_weather_df[f'predicted_{event}'] = model.predict(daily_weather_df[[f'{event}_b', f'{event}_p', 'x_vect', 'y_vect', 'temperature', 'lefty'] + year_inputs + interaction_terms])
            daily_weather_df[f'{event}_factor_{side}'] = daily_weather_df[f'predicted_{event}'] / daily_weather_df[f'{event}_b']

    
    # Select columns to keep
    event_factors_l = [f'{event}_factor_l' for event in events_list]
    event_factors_r = [f'{event}_factor_r' for event in events_list]
    keep = keep_columns + event_factors_l + event_factors_r

    # Set park factors to 1 if venue_id is not among active parks    
    # Check if venue_id is not present in team_map['VENUE_ID']
    not_in_team_map = ~daily_weather_df['venue_id'].isin(team_map['VENUE_ID'].astype(int))
    
    # Set values to 1 for columns in event_factors_l and event_factors_r where the condition is met
    for column in event_factors_l + event_factors_r:
        # Set values to 1 where the condition is met
        daily_weather_df.loc[not_in_team_map, column] = 1


    return daily_weather_df[keep]
```
