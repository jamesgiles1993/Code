```python
complete_dataset = pd.read_csv(os.path.join(baseball_path, "Complete Dataset.csv"))
```

```python
complete_dataset = complete_dataset[['date', 'pitcher', 'pitcherName', 'halfInning', 'inning', 'prePitcherScore', 'preBatterScore', 'startingPitcher']]
```

```python
def assign_leverage(row):
    # If the away team is up, the home team is pitching. They'll guard tie games.
    if row['halfInning'] == "top":
        if row['pitcherName'] == row['startingPitcher']:
            row['Leverage'] = 1
        elif row['inning'] > 6 and -1 <= (row['prePitcherScore'] - row['preBatterScore']) <= 4:
            row['Leverage'] = 3
        elif row['inning'] >= 9 and 0 <= (row['prePitcherScore'] - row['preBatterScore']) <= 3:
            row['Leverage'] = 4
        else:
            row['Leverage'] = 2
    else:
        if row['pitcherName'] == row['startingPitcher']:
            row['Leverage'] = 1
        elif row['inning'] > 6 and -1 <= (row['prePitcherScore'] - row['preBatterScore']) <= 4:
            row['Leverage'] = 3
        elif row['inning'] >= 9 and 1 <= (row['prePitcherScore'] - row['preBatterScore']) <= 3:
            row['Leverage'] = 4
        else:
            row['Leverage'] = 2
```

```python
complete_dataset['Leverage'] = complete_dataset.apply(assign_leverage, axis=1)
```
