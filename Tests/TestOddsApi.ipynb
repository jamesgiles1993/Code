```python
import requests
import json

import pandas as pd
import numpy as np

# Replace 'YOUR_API_KEY' with your actual API key
api_key = 'd26e1a8d59392c1f035a1b9b1db0a1e0'
url = f'https://api.the-odds-api.com/v4/sports/americanfootball_nfl/odds/?apiKey={api_key}&regions=us&markets=h2h,spreads&oddsFormat=american'

# Make a GET request to the API
response = requests.get(url)

# Check if the request was successful (status code 200)
if response.status_code == 200:
    # Parse the JSON data
    data = response.json()

df = pd.json_normalize(data, 'bookmakers', ['id', 'sport_key', 'sport_title', 'commence_time', 'home_team', 'away_team'], sep='_')
```

```python
# Flatten the JSON data into a pandas DataFrame
rows = []
for item in data:
    for bookmaker in item['bookmakers']:
        for market in bookmaker['markets']:
            for outcome in market['outcomes']:
                row = {
                    'id': item['id'],
                    'sport_key': item['sport_key'],
                    'sport_title': item['sport_title'],
                    'commence_time': item['commence_time'],
                    'home_team': item['home_team'],
                    'away_team': item['away_team'],
                    'bookmaker_key': bookmaker['key'],
                    'bookmaker_title': bookmaker['title'],
                    'last_update': bookmaker['last_update'],
                    'market_key': market['key'],
                    'market_last_update': market['last_update'],
                    'outcome_name': outcome['name'],
                    'outcome_price': outcome['price']
                }
                if 'point' in outcome:
                    row['outcome_point'] = outcome['point']
                rows.append(row)

df = pd.DataFrame(rows)

# Display the DataFrame
df
```
