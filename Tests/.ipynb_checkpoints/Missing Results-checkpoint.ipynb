```python
import pandas as pd
import re

# Read the text file
file_path = r"C:\Users\james\Documents\MLB\Data\Leaderboard 20230409.txt"
with open(file_path, "r") as file:
    text_data = file.read()

# print(text_data)
# Extract ranks using regex
ranks = re.findall(r'"rank":(\d+)', text_data)

# Extract entryKeys using regex
entry_keys = re.findall(r'"entryKey":"(\d+)"', text_data)

# Extract userNames using regex
user_names = re.findall(r'"userName":"([^"]+)"', text_data)

# User entry index
entry_index = re.findall(r'"userEntryIndex":(\d+)', text_data)

# Extract fantasyPoints using regex
fantasy_points = re.findall(r'"fantasyPoints":([\d.]+)', text_data)

# Create DataFrame
df = pd.DataFrame({"rank": ranks, "entryKey": entry_keys, "userName": user_names, 'userEntryIndex': entry_index, 'fantasyPoints': fantasy_points})

df = df.iloc[2:].reset_index(drop=True)

# Get actual denominator (won't always be 20)
df['EntryName'] = df['userName'] + " (" + df['userEntryIndex'] + "/" + "20)"

df['TimeRemaining'] = 0
df['Lineup'] = "Missing"

df.rename(columns={'rank':'Rank', 'entryKey':'EntryId', 'fantasyPoints':'Points'}, inplace=True)

df = df[['Rank', 'EntryId', 'EntryName', 'TimeRemaining', 'Points', 'Lineup']]

while df['EntryName'].str.contains('gilesj12').iloc[-1]:
    df = df.iloc[:-1]


df
```
