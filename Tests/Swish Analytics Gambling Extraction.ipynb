```python
import requests
from bs4 import BeautifulSoup
import re
import requests
from bs4 import BeautifulSoup
# Function to scrape tables and extract href numbers from a URL
def scrape_tables_and_extract_href_numbers(url):
    # Send a GET request to the URL
    response = requests.get(url)
    
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')
        
        # Find all tables on the page
        tables = soup.find_all('table')
        
        # Find all href attributes containing "/mlb/game?id="
        href_numbers = re.findall(r'href="/mlb/game\?id=(\d+)"', response.text)
        
        # Create a dictionary to store href numbers corresponding to each table
        table_href_numbers = {}
        
        # Loop through each table and extract its data
        for i, table in enumerate(tables):
            # Initialize an empty list to store rows of the table
            table_data = []
            
            # Find all rows in the table
            rows = table.find_all('tr')
            
            # Loop through each row and extract its cells
            for row in rows:
                # Find all cells in the row
                cells = row.find_all(['th', 'td'])
                
                # Extract text from each cell and append to the row data
                row_data = [cell.get_text(strip=True) for cell in cells]
                
                # Append the row data to the table data
                table_data.append(row_data)
            
            # Print the table data
            print(f"Table {i+1}:")
            for row in table_data:
                print(row)
            print("\n")  # Add a newline between tables
            
            # Associate href numbers with the current table
            if i < len(href_numbers):
                table_href_numbers[table_data[0][0]] = href_numbers[i]
        
        # Print the associated href numbers for each table
        print("Associated href numbers for each table:")
        for table_name, href_number in table_href_numbers.items():
            print(f"Table '{table_name}': {href_number}")
    else:
        print("Failed to retrieve the webpage.")

# URL of the webpage containing tables
url = "https://swishanalytics.com/mlb/today?date=2022-08-23"

# Call the function to scrape tables from the URL
scrape_tables_and_extract_href_numbers(url)

```
