```python
multiplier_dataset = pd.read_csv(r"C:\Users\james\Documents\MLB\Database\Multiplier Dataset.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming multiplier_dataset is your DataFrame
avg_hr_b = multiplier_dataset.groupby('game_date')['hr_wfx_l'].mean()

# Plotting
plt.figure(figsize=(10, 5))
plt.plot(avg_hr_b.index, avg_hr_b.values, marker='o', linestyle='-')
plt.xlabel('Game Date')
plt.ylabel('Average HR_B')
plt.title('Average HR_B per Game Date')
plt.xticks(rotation=45)  # Rotate x-axis labels for readability
plt.grid(True)
plt.show()
```
