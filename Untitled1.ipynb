```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

# Load the dataset
file_path = r"C:\Users\james\Documents\MLB\Database\B03. Lineups\1. Players\Players 175338968.csv"
df = pd.read_csv(file_path)

# Identify FP columns
fp_columns = [col for col in df.columns if col.startswith("FP_")]
info_columns = ["Name", "Position", "AvgPointsPerGame", "TeamAbbrev", "rostership", "roto_projection"]

# Create dropdown widget for player selection
player_dropdown = widgets.Combobox(
    options=df["Name"].unique().tolist(),
    placeholder="Search or select a player",
    description="Player:",
    ensure_option=True
)

output = widgets.Output()

# Function to update output
def update_output(player_name):
    with output:
        output.clear_output()
        player_data = df[df["Name"] == player_name]
        
        if player_data.empty:
            print("Player not found.")
            return

        # Display player info
        display(player_data[info_columns])

        # Plot histogram of FP values
        plt.figure(figsize=(8, 5))
        plt.hist(player_data[fp_columns].values.flatten(), bins=20, edgecolor="black", alpha=0.7)
        plt.xlabel("Fantasy Points")
        plt.ylabel("Frequency")
        plt.title(f"Distribution of Fantasy Points for {player_name}")
        plt.show()

# Bind function to dropdown
def on_change(change):
    if change["new"]:
        update_output(change["new"])

player_dropdown.observe(on_change, names="value")

# Display widgets
display(player_dropdown, output)
```
