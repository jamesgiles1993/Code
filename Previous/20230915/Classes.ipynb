```python
# This creates a class called Batter with all the necessary stats - rates and totals
class Batter:
    def __init__(self, Name, ID, bats, 
                 hbp_l, bb_l, b1_l, b2_l, b3_l, hr_l, so_l, 
                 hbp_r, bb_r, b1_r, b2_r, b3_r, hr_r, so_r, 
                 lo_l, po_l, go_l, fo_l,
                 lo_r, po_r, go_r, fo_r,
                 sba_2b, sba_3b, sb_2b, sb_3b, order
                 ):
        self.Name = Name
        self.ID = ID
        self.bats = bats
        self.bats_copy = bats
        
        # Proclivities against pitcher's hand
        self.hbp = 0
        self.bb = 0
        self.b1 = 0
        self.b2 = 0
        self.b3 = 0
        self.hr = 0
        self.lo = 0
        self.po = 0
        self.go = 0
        self.fo = 0
        self.so = 0
        
        self.sba_2b = 0
        self.sb_2b = 0
        self.sba_3b = 0
        self.sb_3b = 0
        
        # Proclivities vs. lefties
        self.hbp_l = hbp_l
        self.bb_l = bb_l
        self.b1_l = b1_l
        self.b2_l = b2_l
        self.b3_l = b3_l
        self.hr_l = hr_l
        self.lo_l = lo_l
        self.po_l = po_l
        self.go_l = go_l
        self.fo_l = fo_l
        self.so_l = so_l
        
        # Proclivities vs. righties
        self.hbp_r = hbp_r
        self.bb_r = bb_r
        self.b1_r = b1_r
        self.b2_r = b2_r
        self.b3_r = b3_r
        self.hr_r = hr_r
        self.lo_r = lo_r
        self.po_r = po_r
        self.go_r = go_r
        self.fo_r = fo_r
        self.so_r = so_r
        
        # Base running stats
        self.sba_2b = sba_2b
        self.sb_2b = sb_2b
        self.sba_3b = sba_3b
        self.sb_3b = sb_3b
        
        self.order = order
                
        # Game stats    
        self.HBP = 0
        self.BB = 0
        self.B1 = 0
        self.B2 = 0
        self.B3 = 0
        self.HR = 0
        self.SB = 0
        self.R = 0
        self.RBI = 0
        self.FP = 0
```

```python
class Pitcher:
    def __init__(self, Name, ID, throws, 
                 hbp_l, bb_l, b1_l, b2_l, b3_l, hr_l, so_l, 
                 hbp_r, bb_r, b1_r, b2_r, b3_r, hr_r, so_r, 
                 lo_l, po_l, go_l, fo_l,
                 lo_r, po_r, go_r, fo_r,
                 avgouts, starter_sample, leverage
                 ):
        
        self.Name = Name
        self.ID = ID
        self.throws = throws
        
        # Proclivities against batter's hand
        self.hbp = 0
        self.bb = 0
        self.b1 = 0
        self.b2 = 0
        self.b3 = 0
        self.hr = 0
        self.lo = 0
        self.po = 0
        self.go = 0
        self.fo = 0
        self.so = 0

        # Proclivity vs. lefties
        self.hbp_l = hbp_l
        self.bb_l = bb_l
        self.b1_l = b1_l
        self.b2_l = b2_l
        self.b3_l = b3_l
        self.hr_l = hr_l
        self.lo_l = lo_l
        self.po_l = po_l
        self.go_l = go_l
        self.fo_l = fo_l
        self.so_l = so_l

        # Proclivity vs. righties
        self.hbp_r = hbp_r
        self.bb_r = bb_r
        self.b1_r = b1_r
        self.b2_r = b2_r
        self.b3_r = b3_r
        self.hr_r = hr_r
        self.lo_r = lo_r
        self.po_r = po_r
        self.go_r = go_r
        self.fo_r = fo_r
        self.so_r = so_r
        
        # Pitcher stats
        self.leverage = leverage
        self.avgouts = avgouts
        self.starter_sample = starter_sample
        
        # Pitcher game stats
        self.winning = False
        self.OUT = 0
        self.SO = 0
        self.HBP = 0
        self.BB = 0
        self.B1 = 0
        self.B2 = 0
        self.B3 = 0
        self.HR = 0
        self.H = 0
        self.CS = 0
        self.ER = 0
        self.W = 0
        self.CG = 0
        self.CGSO = 0
        self.NH = 0
        self.FP = 0
        self.PA = 0
```
