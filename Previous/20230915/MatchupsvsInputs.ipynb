```python
import os
import pandas as pd

# Specify the path to the directory
directory_path = r'C:\Users\james\Documents\MLB\Database\B01. Matchups'

# Initialize an empty list to store DataFrames
dfs = []

# Iterate through all folders and files within the specified directory
for root, dirs, files in os.walk(directory_path):
    # Filter out folders that do not start with "Matchups 2022"
    dirs[:] = [d for d in dirs if d.startswith('Matchups 2022')]
    
    for filename in files:
        if filename.endswith('.xlsx'):
            file_path = os.path.join(root, filename)
            
            # Read Excel file
            xls = pd.ExcelFile(file_path)
            
            # Read data from the desired sheets
            away_batters = pd.read_excel(xls, 'AwayPitchers')
            home_batters = pd.read_excel(xls, 'HomePitchers')
            
            # Filter rows where "batting_order" is not missing
            away_batters = away_batters.dropna(subset=['Leverage'])
            home_batters = home_batters.dropna(subset=['Leverage'])
            
            # Append DataFrames to the list
            dfs.append(away_batters)
            dfs.append(home_batters)
```

```python
# ncatenate all DataFrames in the list
result_df = pd.concat(dfs, ignore_index=True)

result_df = result_df.query('date >= 20220507 and date <= 20221011')

# Display the result DataFrame
result_df.describe()
```
