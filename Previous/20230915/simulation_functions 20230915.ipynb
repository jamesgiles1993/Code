```python
def create_game(matchup, matchup_path, team_map, weather_df):
    AwayBatters, HomePitchers, HomeBatters, AwayPitchers, park = create_matchup(matchup, matchup_path)
    venue = team_map.loc[park, 'VENUE_ID']
    game = Scoreboard(AwayBatters, HomePitchers, HomeBatters, AwayPitchers, park, 9)

    venue_id = "venue_" + str(venue)
    
    # Loop over all the variables in the game object
    for attr, value in vars(game).items():
        # If the attribute is the venue ID
        if attr == venue_id:
            # Set it to 1
            value = 1
            setattr(game, attr, value)
        
    game.TEMP_PARK_CT = weather_df.loc[venue]['TEMP_PARK_CT']
    game.x_vect = weather_df.loc[venue]['x_vect']
    game.y_vect = weather_df.loc[venue]['y_vect']
    
    return game
```

```python
# Hit by pitch
def hbp(game):
    # Runs
    runs = 0
    
    # Players
    game.pitching.HBP += 1
    game.ab.HBP += 1 
    
    # Bases: 1st, 2nd, 3rd
    if game.on_3b is not None and game.on_2b is not None and game.on_1b is not None:
        game.on_3b.R += 1
        game.pitching.ER += 1
        game.ab.RBI += 1
        runs += 1
        
        game.on_3b = game.on_2b
        game.on_2b = game.on_1b
        game.on_1b = game.ab
        
    # Bases: 1st, 2nd
    elif game.on_3b is None and game.on_2b is not None and game.on_1b is not None:
        game.on_3b = game.on_2b
        game.on_2b = game.on_1b
        game.on_1b = game.ab
        
    # Bases: 1st, 3rd
    elif game.on_3b is not None and game.on_2b is None and game.on_1b is not None:
        game.on_3b = game.on_3b
        game.on_2b = game.on_1b
        game.on_1b = game.ab
        
    # Bases: 2nd, 3rd
    elif game.on_3b is not None and game.on_2b is not None and game.on_1b is None:
        game.on_3b = game.on_3b
        game.on_2b = game.on_2b
        game.on_1b = game.ab
    
    # Bases: 1st
    elif game.on_3b is None and game.on_2b is None and game.on_1b is not None:
        game.on_3b = None
        game.on_2b = game.on_1b
        game.on_1b = game.ab
        
    # Bases: 2nd
    elif game.on_3b is None and game.on_2b is not None and game.on_1b is None:
        game.on_3b = None
        game.on_2b = game.on_2b
        game.on_1b = game.ab
    
    # Bases: 3rd
    elif game.on_3b is not None and game.on_2b is None and game.on_1b is None:
        game.on_3b = game.on_3b
        game.on_2b = None
        game.on_1b = game.ab
    
    # Bases: None
    elif game.on_3b is None and game.on_2b is None and game.on_1b is None:
        game.on_3b = None
        game.on_2b = None
        game.on_1b = game.ab
    
    # Runs scored
    if game.top_bot == "Top":
        game.away_score += runs
    else:
        game.home_score += runs
                
    return game
```

```python
# Out in play
# Add sac fly support
def in_play_out(game, out, dp_rates, dp_base_rates, advances):
    # Runs
    runs = 0
    
    game.pitching.OUT += 1
    game.outs += 1
    
    # Create flags for bases occupied. Use to index
    if game.on_1b is not None:
        runner_1b = 1
    else:
        runner_1b = 0
    if game.on_2b is not None:
        runner_2b = 1
    else:
        runner_2b = 0
    if game.on_3b is not None:
        runner_3b = 1
    else: 
        runner_3b = 0

    # Type of double play it would be
    out_dp = out + "_dp"
    
    # Odds that it's a double play given that it's that type of out
    dp_odds = dp_rates[out_dp].loc[(runner_1b, runner_2b, runner_3b)]

    
    # Double play odds are not conditional on there being an out. Make 'em so.
    if out == "ld":
        dp_odds = dp_odds / game.ab.lo
    elif out == "pu":
        dp_odds = dp_odds / game.ab.po
    elif out == "go":
        dp_odds = dp_odds / game.ab.go
    elif out == "fo":
        dp_odds = dp_odds / game.ab.fo
    
    # Is it a double play?
    dp_roll = random.random() 
    if dp_roll < dp_odds:
        # print("double play!")
        game.outs += 1
        game.pitching.OUT += 1
        
        # Second out 
        # At 1B
        dp_1b = out_dp + "_1b"
        dp_1b_odds = dp_base_rates[dp_1b].loc[(runner_1b, runner_2b, runner_3b)]
        
        # Ground outs are weird.
        if out == "go":
            dp_1b_odds = 0
        
        # At 2B
        dp_2b = out_dp + "_2b"
        dp_2b_odds = dp_base_rates[dp_2b].loc[(runner_1b, runner_2b, runner_3b)]
        
        # At 3B
        dp_3b = out_dp + "_3b"
        dp_3b_odds = dp_base_rates[dp_3b].loc[(runner_1b, runner_2b, runner_3b)] + dp_2b_odds
        
        # At home
        dp_h = out_dp + "_h"
        dp_h_odds = dp_base_rates[dp_h].loc[(runner_1b, runner_2b, runner_3b)] + dp_3b_odds
        
        # Two on the bases (only matters for GOs)
        dp_b = out_dp + "_b"
        dp_b_odds = dp_base_rates[dp_b].loc[(runner_1b, runner_2b, runner_3b)] + dp_h_odds
        
        # If the out is NOT a ground out!
        if out != "go":
            # Normalize by multiplying by max because sometimes it gets a little off
            dp_base_roll = random.random() * dp_b_odds
            # If it's an out at first
            if dp_base_roll < dp_1b_odds:
                # And there's nobody at first
                if game.on_1b is None:
                    # It's an out at second
                    game.on_2b = None
                # Else, it's just an out at first
                game.on_1b = None
            # If it's an out at second
            elif dp_base_roll < dp_2b_odds:
                # And there's nobody on second
                if game.on_2b is None:
                    # It's an out at first
                    game.on_1b = None
                game.on_2b = None
            # If it's an out at third
            elif dp_base_roll < dp_3b_odds:
                # And there's nobody on third or second
                if game.on_3b is None and game.on_2b is None:
                    # It's an out at first
                    game.on_1b = None
                # If there's nobody on third but there is on second 
                elif game.on_3b is None: 
                    # It's an out at second
                    game.on_2b = None
                game.on_3b = None   
            # If it's an out at home
            elif dp_base_roll < dp_h_odds:
                # It's an out at third
                game.on_3b = None
            # Can't be two outs on bases
            
        # If the out is a ground out
        if out == "go": 
            # Normalize by multiplying by max because sometimes it gets a little off
            dp_base_roll = random.random() * dp_b_odds
            # These are shifted to affect the runner 
            if dp_base_roll < dp_2b_odds:
                if game.on_1b is None:
                    game.on_2b = None
                game.on_1b = None
            elif dp_base_roll < dp_3b_odds:
                if game.on_2b is None and game.on_3b is None:
                    game.on_1b = None
                elif game.on_2b is None:
                    game.on_3b = None
                game.on_2b = None    
            elif dp_base_roll < dp_h_odds:
                game.on_3b = None 
            # There are two outs on the bases
            else:
                # If there are runners on second and third
                if game.on_1b is None and game.on_2b is not None and game.on_3b is not None:
                    # There are outs at second and third
                    game.on_2b = None
                    game.on_3b = None
                # If there are runners on first and third
                elif game.on_1b is not None and game.on_2b is None and game.on_3b is not None:
                    # There are outs at first and third
                    game.on_1b = None
                    game.on_3b = None
                # If there are runners on first and second
                elif game.on_1b is not None and game.on_2b is not None and game.on_3b is None:
                    # There are outs at first and second
                    game.on_1b = None
                    game.on_2b = None
                # If there are bases loaded
                elif game.on_1b is not None and game.on_2b is not None and game.on_3b is not None:
                    # There are outs at second and third (this is a simplifying assumption)
                    game.on_2b = None
                    game.on_3b = None
                # Batter reaches                
                game.on_1b = game.ab       
        dp = 1
        
    else:
        dp = 0
    
    # Advances 
    if game.outs < 3:
        if game.on_3b is not None:
            # Roll to see if they'll score
            advances_h_roll = random.random()
            advance_type = out + "_3b_h"
            if advances_h_roll < advances[advance_type][dp]: 
                game.on_3b.R += 1
                game.pitching.ER += 1
                game.on_3b = None
                if dp == 0:
                    game.ab.RBI += 1
                runs += 1
#                 print("Home")
                    
        if game.on_2b is not None and game.on_3b is None:
            # Roll to see if they'll go to third
            advances_3b_roll = random.random()
            advance_type = out + "_2b_3b"
            if advances_3b_roll < advances[advance_type][dp]: 
                game.on_3b = game.on_2b 
                game.on_2b = None
#                 print("Third")
        
                    
        if game.on_1b is not None and game.on_2b is None:
            # Roll to see if they'll go to second
            advances_2b_roll = random.random()
            advance_type = out + "_1b_2b"
            if advances_2b_roll < advances[advance_type][dp]: 
                game.on_2b = game.on_1b 
                game.on_1b = None
#                 print("Second")
            
    # Runs score
    if game.top_bot == "Top":
        game.away_score += runs
    else:
        game.home_score += runs

    return game    
```

Double play base explainer:
Double play base rates takes base runner locations at the beginning of the PA and determines the likelihood of outs at each base or the odds of there being two outs on the bases (so the batter reaches first).
This is conditional on it being a double play.
For all outs but groundouts, you can basically just take the guy on the randomly selected base and remove him
So if there are guys on second and third and the second out is on third, remove the guy from third (this is a simplifying assumption, but it's also not really a big deal)
A few times, this gets a little weird so there are exceptions made.
If there is nobody on second base, for example, and that's where the out is, you can assume a runner tried going from first, so call him out.
For ground outs, the base level assumption is that runners are out at the base they are advancing to, not the one they're currently at.
So an out at second removes the runner at first by default.
This also has some weird exceptions, including a tiny likelihood of a guy at first getting out at third, but this is rare (and probably could be fixed by rounding)

```python
# Calculate batter fantasy points
def calculate_batter(batter, game):
    batter.FP = (
                batter.B1 * 3 +
                batter.B2 * 5 +
                batter.B3 * 8 +
                batter.HR * 10 +
                batter.RBI * 2 +
                batter.R * 2 +
                batter.BB * 2 +
                batter.HBP * 2 +
                batter.SB * 5
                )
    
    return batter, game
```

```python
# Calculate pitcher fantasy points
def calculate_pitcher(pitcher, game):
    # Calculate hits allowed
    pitcher.H = (pitcher.B1 + pitcher.B2 + pitcher.B3 + pitcher.HR)

    # If they're the winning pitcher, they get a win
    if game.winning_pitcher == pitcher and game.inning >= game.innings:
        pitcher.W = 1
    else:
        pitcher.W = 0

    # Determine CG, CGSO, NH
    if pitcher.OUT == 27:
        pitcher.CG = 1
        if pitcher.ER == 0:
            pitcher.CGSO = 1
        if pitcher.H == 0:
            pitcher.NH = 1

    pitcher.FP = (
                pitcher.OUT * 0.75 +
                pitcher.SO * 2 +
                pitcher.W * 4 +
                pitcher.ER * -2 +
                pitcher.H * -0.6 +
                pitcher.BB * -0.6 +
                pitcher.HBP * -0.6 +
                pitcher.CG * 2.5 +
                pitcher.CGSO * 2.5 +
                pitcher.NH * 5
                )
    
    pitcher.PA = pitcher.OUT + pitcher.H + pitcher.BB + pitcher.HBP
    
    
    return pitcher, game
```

```python
# Calculate odds of being pulled
def pull_odds(game, pull):
    # Determine batting and fielding team scores for use as inputs
    if game.top_bot == "Top":        
        bat_score = game.away_score
        fld_score = game.home_score
    else:
        bat_score = game.home_score
        fld_score = game.away_score
    
    # Pull model inputs
    X_list = [
        game.pitching.B1, game.pitching.B2, game.pitching.B3, game.pitching.BB, game.pitching.HBP, game.pitching.HR, 
        game.pitching.SO, game.pitching.faced, game.pitching.ER, fld_score, bat_score, game.inning, game.outs, game.pitching.starter, game.pitching.avgFaced
            ]
    
    # Reshape to match what models expect
    model_inputs = pd.Series(X_list).values.reshape(1,-1)

    # Predict and make into a list
    pull_list = pull.predict_proba(model_inputs).tolist()
    
    # Choose the probability that pull = 1
    odds = pull_list[0][1] 
    
    del X_list, model_inputs, pull_list
    
    return odds
```

```python
def ab(game, hbp, bb, b1, b2, b3, hr, so, lo, po, go, fo):
    # Roll
    pa_roll = random.random() 
    
    # Event 1: HBP
    if pa_roll < hbp:
        event = "HBP"
    # Event 2: BB
    elif pa_roll < bb:
        event = "Walk"
    # Event 3: Single
    elif pa_roll < b1:
        event = "Single"
    # Event 4: Double    
    elif pa_roll < b2:
        event = "Double"
    # Event 5: Triple
    elif pa_roll < b3:
        event = "Triple"
     # Event 6: Home Run
    elif pa_roll < hr:
        event = "Home Run"
    # Event 7: Strikeout
    elif pa_roll < so:
        event = "Strikeout"
    # Event 8: Line drive out
    elif pa_roll < lo:
        event = "ld"
    # Event 9: Popup out
    elif pa_roll < po:
        event = "pu"
    # Event 10: Ground ball out
    elif pa_roll < go:
        event = "go"
    # Event 11: Fly-out 
    else:
        event = "fo"
        
    return game, event
```

```python
def sim_ab(game, pull, model1, model2, model3, model4, opener_list):
    # Determine home pitcher
    # Shuffle list of home pitchers so that when a pitcher is decided based on being the first in the list with a given leverage, that pitcher will vary
    random.shuffle(game.home_pitchers)
    try:
        game.home_pitcher_up = next(pitcher for pitcher in game.home_pitchers if pitcher.leverage == game.home_pitcher)
    except:
        game.home_pitcher_up = next(pitcher for pitcher in game.home_pitchers if pitcher.leverage == 2)

    # Determine away pitcher
    random.shuffle(game.away_pitchers)
    try:
        game.away_pitcher_up = next(pitcher for pitcher in game.away_pitchers if pitcher.leverage == game.away_pitcher)
    except:
        game.away_pitcher_up = next(pitcher for pitcher in game.away_pitchers if pitcher.leverage == 2)
    
    
    
    # Determine matchup       
    if game.top_bot == "Top":
        game.ab = next(batter for batter in game.away_batters if batter.order == game.away_batter)
        game.pitching = game.home_pitcher_up
    else:
        game.ab = next(batter for batter in game.home_batters if batter.order == game.home_batter)
        game.pitching = game.away_pitcher_up
        
    if game.pitching.leverage == 1:
        game.pitching.starter = 1
    
    if game.top_bot == "Top" and game.home_starter_pulled == False:
        home_pull = pull_odds(game, pull)
        pull_prob = random.random()
        if pull_prob < home_pull:
            game.home_starter_pulled = True    
            # print("Home Starter Pulled in the " + str(game.inning))
    
    elif game.top_bot == "Bot" and game.away_starter_pulled == False:
        away_pull = pull_odds(game, pull)
        pull_prob = random.random()
        if pull_prob < away_pull:
            game.away_starter_pulled = True  
            # print("Away Starter Pulled in the " + str(game.inning))   
            # print(game.home_score)
            
    if game.inning >= 3 and game.home_starter.Name in opener_list:
        game.home_starter_pulled = True
    if game.inning >= 3 and game.away_starter.Name in opener_list:
        game.away_starter_pulled = True
    
    # Determine a leverage situation (1 is SP)
    # If the starter is pulled
    if game.away_starter_pulled == True:
        game.away_pitcher = 2
    if game.home_starter_pulled == True:
        game.home_pitcher = 2

    # If it's the 7th or later
    if game.inning > game.innings - 3:
        if game.away_starter_pulled == True:
             game.away_pitcher = 3
        if game.home_starter_pulled == True:
            game.home_pitcher = 3

    # If it's a save/tie situation in the 9th
    if (game.inning == game.innings):
        if (4 > (game.away_score - game.home_score) >= 0) and game.away_starter_pulled == True:
            game.away_pitcher = 4
        if (4 > (game.home_score - game.away_score) >= 0) and game.home_starter_pulled == True:
            game.home_pitcher = 4
    



    # Determine event probabilities
    hbp, bb, b1, b2, b3, hr, so, lo, po, go, fo, game = probabilities(game, model1, model2, model3, model4)
    # Determine event
    game, event = ab(game, hbp, bb, b1, b2, b3, hr, so, lo, po, go, fo)
    
    game.pitching.faced += 1
    
    return game, event
```

```python
def sim_inning(game, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, opener_list, innings=9):
    # Inning
#     print(game.top_bot + " " + str(game.inning))

    # Set outs to zero
    game.outs = 0
 
    # Loop for each PA
    while game.outs < 3:
        # Set runs for that PA at 0 (not sure this actually does anything - also set in each event)
        runs = 0
        
        # Steals
        # Third base
        # If third is empty and second is not
        if game.on_3b is None and game.on_2b is not None:
            sba_3b_roll = random.random()
            # Attempt to steal third
            if sba_3b_roll < game.on_2b.sba_3b :
                sb_3b_roll = random.random()
                # They steal third
                if sb_3b_roll < game.on_2b.sb_3b:
                    game.on_2b.SB += 1
                    game.on_3b = game.on_2b
                    game.on_2b = None
                    
                # They're out
                else:
                    game.on_2b = None
                    game.outs += 1 
                    
        # Second base
        # If second is empty and first is not
        if game.on_2b is None and game.on_1b is not None:
            sba_2b_roll = random.random()
            # Attempt to steal second
            if sba_2b_roll < game.on_1b.sba_2b :
                sb_2b_roll = random.random()
                # They steal second
                if sb_2b_roll < game.on_1b.sb_2b:
                    game.on_1b.SB += 1
                    game.on_2b = game.on_1b
                    game.on_1b = None
                    
                # They're out
                else:
                    game.on_1b = None
                    game.outs += 1
                    
        
        
        # Which team is winning?
        if game.away_score > game.home_score:
            game.winning_team = "Away"
        elif game.home_score > game.away_score:
            game.winning_team = "Home"
        elif game.home_score == game.away_score:
            game.winning_team = "Tie"
        
        # Who gets the win?
        # If the away team is winning
        if game.winning_team == "Away":
            # And it's the fifth inning or later (pitchers finish their innings, so this is fine)
            if game.inning >= 5:
                # And the current winning pitcher is not on the away team (None or on the other team)
                if game.winning_pitcher not in game.away_pitchers:
                    # The current pitcher is the winning pitcher
                    game.winning_pitcher = game.away_pitcher_up
        elif game.winning_team == "Home":
            if game.inning >= 5:
                if game.winning_pitcher not in game.home_pitchers:
                    game.winning_pitcher = game.home_pitcher_up
        else:
            game.winning_pitcher = None
            
#         print("Current winning pitcher")
#         try:
#             print(game.winning_pitcher.Name)
#         except:
#             print("None yet")
            
        
        # Simulate the AB
        game, event = sim_ab(game, pull, model1, model2, model3, model4, opener_list)
        # print(event)

        # Determine the implications of the event
        # Strikeouts have no implications
        if event == "Strikeout":
            game.outs += 1
            game.pitching.SO += 1
            game.pitching.OUT += 1

        elif event == "HBP":
            game = hbp(game)
            
        elif event == "Walk":
            game = walk(game)
            
        elif event == "Single":
            game = single(game)
            
        elif event == "Double":
            game = double(game)
            
        elif event == "Triple":
            game = triple(game)
            
        elif event == "Home Run":
            game = home_run(game)
            
        else:
            game = in_play_out(game, event, dp_rates, dp_base_rates, advances)
        
        # Go to the next batter up
        if game.top_bot == "Top":
            game.away_score += runs
            game.away_batter += 1
            if game.away_batter == 10:
                game.away_batter = 1

        else:
            game.home_score += runs
            game.home_batter += 1
            if game.home_batter == 10:
                game.home_batter = 1



        # print number of outs
        # print(str(game.outs) + " Outs")

        # print who's on base
        # if game.on_1b != None:
            # print("On first: " + game.on_1b.Name)
        # if game.on_2b != None:
            # print("On second: " + game.on_2b.Name)
        # if game.on_3b != None:
            # print("On third: " + game.on_3b.Name)

        # If it's the final inning and the bottom of it
        if game.inning == innings and game.top_bot == "Bot":
            # At any point, if the home team is winning, it's over
            if game.home_score > game.away_score:
                game.outs = 3
        
    # Reset bases after inning
    game.on_1b = None
    game.on_2b = None
    game.on_3b = None
    
    game.onFirst = None
    game.onSecond = None
    game.onThird = None

    # Transition the inning
    if game.top_bot == "Top":
        game.top_bot = "Bot"
    elif game.top_bot == "Bot":
        game.top_bot = "Top"
        game.inning += 1

#     print score
#     print("Away: " + str(game.away_score))
#     
# : " + str(game.home_score))
    del event

    return game
```

```python
# Simulate a whole game
def sim_game(game, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, opener_list):
    park = game.park
    innings = game.innings
    
    # Determine starters
    game.home_starter = next(pitcher for pitcher in game.home_pitchers if pitcher.leverage == 1)
    game.away_starter = next(pitcher for pitcher in game.away_pitchers if pitcher.leverage == 1)
    
    # Loop over every inning
    while game.inning <= innings and innings <= 15:
        # print(game.inning)
        # Calculate a pitcher's cumulative stats
        game.home_starter, game = calculate_pitcher(game.home_starter, game)
        game.away_starter, game = calculate_pitcher(game.away_starter, game)
        
        # Simulate the half inning  
        game = sim_inning(game, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, opener_list, innings)

        # If it's the bottom of the 9th and the home team is winning
        if (game.top_bot == "Bot") and (game.inning >= innings) and (game.home_score > game.away_score):
            # print("Game Over")
            
            for batter in game.home_batters:
                batter, game = calculate_batter(batter, game)
            for batter in game.away_batters:
                batter, game = calculate_batter(batter, game)
            for pitcher in game.home_pitchers:
                pitcher, game = calculate_pitcher(pitcher, game)
            for pitcher in game.away_pitchers:
                pitcher, game = calculate_pitcher(pitcher, game)

            return game

        # Else, simulate another half inning
        else:
            game = sim_inning(game, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, opener_list, innings)
        
        if game.away_score == game.home_score and game.inning == innings + 1 and game.top_bot == "Top":
            # print("TIE")
            innings += 1         
        
        
    for batter in game.home_batters:
        batter, game = calculate_batter(batter, game)
    for batter in game.away_batters:
        batter, game = calculate_batter(batter, game)
    for pitcher in game.home_pitchers:
        pitcher, game = calculate_pitcher(pitcher, game)
    for pitcher in game.away_pitchers:
        pitcher, game = calculate_pitcher(pitcher, game)

    # Doesn't currently pick winning pitchers of walkoffs for some reason, probably not a big deal, but fix
    return game
```

```python
def create_score_df(game):
    # Away Batters
    name_list = []
    ID_list = []
    order_list = []
    FP_list = []
    for batter in game.away_batters:
        name_list.append(batter.Name)
        ID_list.append(batter.ID)
        order_list.append(batter.order)
        FP_list.append(batter.FP)

    away_batter_df = pd.DataFrame(list(zip(name_list, ID_list, order_list, FP_list)), columns=['Name', 'ID', 'order', 'FP'])
    away_batter_df = away_batter_df.query("ID != 0")

    # Home Batters
    name_list = []
    ID_list = []
    order_list = []
    FP_list = []
    for batter in game.home_batters:
        name_list.append(batter.Name)
        ID_list.append(batter.ID)
        order_list.append(batter.order)
        FP_list.append(batter.FP)

    home_batter_df = pd.DataFrame(list(zip(name_list, ID_list, order_list, FP_list)), columns=['Name', 'ID', 'order', 'FP'])
    home_batter_df = home_batter_df.query("ID != 0")

    # Away Pitchers
    name_list = []
    ID_list = []
    FP_list = []
    for pitcher in game.away_pitchers:
        name_list.append(pitcher.Name)
        ID_list.append(pitcher.ID)
        FP_list.append(pitcher.FP)

    away_pitcher_df = pd.DataFrame(list(zip(name_list, ID_list, FP_list)), columns=['Name', 'ID', 'FP'])
    away_pitcher_df = away_pitcher_df.query("ID != 0")

    # Home Pitchers
    name_list = []
    ID_list = []
    FP_list = []
    for pitcher in game.home_pitchers:
        name_list.append(pitcher.Name)
        ID_list.append(pitcher.ID)
        FP_list.append(pitcher.FP)

    home_pitcher_df = pd.DataFrame(list(zip(name_list, ID_list, FP_list)), columns=['Name', 'ID', 'FP'])
    home_pitcher_df = home_pitcher_df.query("ID != 0")
    
    score_df = away_batter_df.append(home_batter_df)
    score_df = score_df.append(away_pitcher_df)
    score_df = score_df.append(home_pitcher_df, ignore_index=True)
    
    del away_batter_df, home_batter_df, away_pitcher_df, home_pitcher_df, name_list, ID_list, FP_list
    return score_df
```

```python
def sim_matchup(matchup, matchup_path, team_map, weather_df, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, 
                opener_list, all_scores_list, game_template):
    
    # New game
    game = deepcopy(game_template)
    game = sim_game(game, pull, model1, model2, model3, model4, dp_rates, dp_base_rates, advances, opener_list)
    score_df = create_score_df(game)

    away_score = game.away_score
    home_score = game.home_score
    
    # Only keep one Ohtani
    # Calculate average score
    points_cols = [col for col in score_df.columns if 'FP' in col]
    score_df['AvgPointsPerGame'] = score_df[points_cols].mean(axis=1)
    # Sort by ID and AvgPointsPerGame, highest score will be on the bottom
    score_df.sort_values(by=['ID', 'AvgPointsPerGame'], ascending=True, inplace=True, ignore_index=True)
    # Keep that one. Will always be hitting stats when he doesn't pitch. 
    # If he does pitch, it'll probably be his pitching stats, although a better solution is probably possible.
    score_df.drop_duplicates(subset='ID', keep='last', inplace=True)
    # Get rid of this column as it will interfere later
    score_df.drop(columns={'AvgPointsPerGame'}, inplace=True)
    
    return score_df, away_score, home_score
```
