```python
def odds(date):
    # Convert to proper format
    date_dash = date[0:4] + "-" + date[4:6] + "-" + date[6:8]
    
    url = f'https://www.fantasylabs.com/api/sportevents/3/{date_dash}/vegas/?sportId=3&date={date_dash}'
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
    jsonData = requests.get(url, headers=headers).json()
    
    # Initialize an empty list to store extracted data
    data_list = []

    for game in jsonData:
        event_details = game.get('EventDetails', {})

        if event_details == {'Properties': {'Scores': []}, 'IsLive': False}:
            continue

        properties = event_details.get('Properties', {})

        VisitorTeamShort = properties.get('VisitorTeamShort')
        HomeTeamShort = properties.get('HomeTeamShort')
        Spread = properties.get('HomeGameSpreadCurrent')
        OU = properties.get('VisitorGameOUCurrent')
        SpreadMoney1 = properties.get('VisitorGameSpreadJuiceCurrent')
        SpreadMoney2 = properties.get('HomeGameSpreadJuiceCurrent')
        OuMoney1 = properties.get('VisitorGameOUJuiceCurrent')
        OuMoney2 = properties.get('HomeGameOUJuiceCurrent')
        MLMoney1 = properties.get('VisitorGameMoneylineCurrent')
        MLMoney2 = properties.get('HomeGameMoneylineCurrent')
        VisitorVegasRuns = properties.get('VisitorVegasRuns')
        HomeVegasRuns = properties.get('HomeVegasRuns')
        EventDateTime = game.get('EventDateTime')

        # Append extracted data to the list
        data_list.append({
            'VisitorTeamShort': VisitorTeamShort,
            'HomeTeamShort': HomeTeamShort,
            'Spread': Spread,
            'OU': OU,
            'SpreadMoney1': SpreadMoney1,
            'SpreadMoney2': SpreadMoney2,
            'OuMoney1': OuMoney1,
            'OuMoney2': OuMoney2,
            'MLMoney1': MLMoney1,
            'MLMoney2': MLMoney2,
            'VisitorVegasRuns': VisitorVegasRuns,
            'HomeVegasRuns': HomeVegasRuns,
            'EventDateTime': EventDateTime,
            'date': date
        })


    # Create a Pandas DataFrame from the list
    odds_df = pd.DataFrame(data_list)

    return odds_df
```
