```python
import os
import re
import pandas as pd
```

## 1. Results

### Contest Results

```python
folder_path = r"C:\Users\james\Documents\MLB\Database\A01. DraftKings\4. Results"

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Regular expression pattern to match the numbers
pattern = r"contest-standings-(\d+)"

# List to store the extracted numbers
resultKeys = []

# Iterate over the files and extract the numbers
for file in files:
    match = re.search(pattern, file)
    if match:
        contestKey = match.group(1)
        resultKeys.append(contestKey)
        
len(resultKeys)
```

## 2. Contests

### Payouts

```python
folder_path = r"C:\Users\james\Documents\MLB\Database\A01. DraftKings\3. Payouts"

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Regular expression pattern to match the numbers
pattern = r"Payouts (\d+)"

# List to store the extracted numbers
payoutKeys = []

# Iterate over the files and extract the numbers
for file in files:
    match = re.search(pattern, file)
    if match:
        contestKey = match.group(1)
        payoutKeys.append(contestKey)
        
len(payoutKeys)
```

### Salaries

```python
folder_path = r"C:\Users\james\Documents\MLB\Database\A01. DraftKings\2. Draftables"

# Get a list of all files in the folder
files = os.listdir(folder_path)

# Regular expression pattern to match the numbers
pattern = r"Draftables (\d+)"

# List to store the extracted numbers
salaryKeys = []

# Iterate over the files and extract the numbers
for file in files:
    match = re.search(pattern, file)
    if match:
        contestKey = match.group(1)
        salaryKeys.append(contestKey)
        
len(salaryKeys)
```

```python
# Create a set of all unique contest keys
contestKeys = set(resultKeys + payoutKeys + salaryKeys)

# Create a dataframe with contestKey column
df = pd.DataFrame({'contestKey': list(contestKeys)})

# Add columns indicating the presence of contestKey in each list
df['result'] = df['contestKey'].isin(resultKeys).astype(int)
df['payout'] = df['contestKey'].isin(payoutKeys).astype(int)
df['salary'] = df['contestKey'].isin(salaryKeys).astype(int)
```

```python
import os
import pandas as pd
from dateutil import parser

salaries_folder_path = r"C:\Users\james\Documents\MLB\Database\A01. DraftKings\2. Draftables"
payouts_folder_path = r"C:\Users\james\Documents\MLB\Database\A01. DraftKings\3. Payouts"

# Create new columns in the existing dataframe
df['date'] = None
df['entryFee'] = None
df['entries'] = None
df['draftGroupId'] = None
df['name'] = None

# Loop over values in "contestKey" column
for index, row in df.iterrows():
    contestKey = row['contestKey']
    
    # Read Salaries CSV file
    salaries_file_path = os.path.join(salaries_folder_path, f"Draftables {contestKey}.csv")
    if os.path.isfile(salaries_file_path):
        df_salaries = pd.read_csv(salaries_file_path, encoding='iso-8859-1')
        
        # Extract the date from the first row of "Game Info" column
        first_row_game_info = df_salaries.loc[0, 'Game Info']
        date_match = re.search(r'\d{2}/\d{2}/\d{4}', first_row_game_info)
        if date_match:
            date_str = date_match.group()
            date = parser.parse(date_str).strftime('%Y%m%d')
            df.at[index, 'date'] = date
    
    # Read Payouts CSV file
    payouts_file_path = os.path.join(payouts_folder_path, f"Payouts {contestKey}.csv")
    if os.path.isfile(payouts_file_path):
        df_payouts = pd.read_csv(payouts_file_path, encoding='iso-8859-1')
        
        # Extract desired columns from the first observation
        if not df_payouts.empty:
            df.at[index, 'entryFee'] = df_payouts.loc[0, 'entryFee']
            df.at[index, 'entries'] = df_payouts.loc[0, 'entries']
            df.at[index, 'draftGroupId'] = df_payouts.loc[0, 'draftGroupId']
            df.at[index, 'name'] = df_payouts.loc[0, 'name']
```
