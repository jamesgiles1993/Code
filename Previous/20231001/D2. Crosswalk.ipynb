```python
def read_position(position):
    # Define the folder path where the files are located (change this to your Downloads folder path)
    folder_path = r"C:\Users\james\Downloads"

    # List all files in the folder that start with "steamer_{position}"
    file_list = [file for file in os.listdir(folder_path) if file.startswith(f'steamer_{position}')]

    if not file_list:
        raise FileNotFoundError(f"No files found in {folder_path} starting with 'steamer_{position}'.")

    # Find the most recent file based on file creation time
    most_recent_file = max(file_list, key=lambda f: os.path.getctime(os.path.join(folder_path, f)))
    
    print(most_recent_file)

    # Read the most recent file into a DataFrame
    df = pd.read_csv(os.path.join(folder_path, most_recent_file))

    # Convert the 'proj_date' column to a datetime object
    df['proj_date'] = pd.to_datetime(df['proj_date'], format='%Y-%m-%d')

    # Filter rows where proj_date is 2021 or later
    df = df[df['proj_date'].dt.year >= 2021]
    
    # Keep the most recent instance of unique combinations of mlbamid and steamerid
    df = df.sort_values('proj_date', ascending=False).drop_duplicates(['mlbamid', 'steamerid'])
    
    # Keep only the specified columns
    df = df[['mlbamid', 'steamerid', 'firstname', 'lastname']]



    return df
```

```python
def create_xwalk():
    download_logs()
    
    batter_xwalk = read_position('hitters')
    pitcher_xwalk = read_position('pitchers')

    xwalk = pd.concat([batter_xwalk, pitcher_xwalk], axis=0)
    
    xwalk['mlbamid'] = xwalk['mlbamid'].astype('int')
    
    xwalk.to_csv(os.path.join(baseball_path, "Utilities", "Player Crosswalk.csv"), encoding='iso-8859-1', index=False)
    
    return xwalk
```
