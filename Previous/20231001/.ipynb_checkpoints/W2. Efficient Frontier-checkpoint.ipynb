```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt


baseball_path = r"C:\Users\james\Documents\MLB\Data"
```

```python
df = pd.read_csv(os.path.join(baseball_path, "A9. Optimizer", "Lineup_Sims_20230601.csv"))

# Compute the mean and standard deviation for each row
df['Sim Mean'] = df[df.columns[df.columns.str.startswith('Sim')]].mean(axis=1)
df['Sim STD'] = df[df.columns[df.columns.str.startswith('Sim')]].std(axis=1)
```

```python
import matplotlib.pyplot as plt

# Assuming your DataFrame is named 'df'
x = df['Sim STD']
y = df['Sim Mean']

# Create a scatter plot
plt.scatter(x, y)

# Set labels and title
plt.xlabel('Sim STD')
plt.ylabel('Sim Mean')
plt.title('Scatter Plot: Sim STD vs. Sim Mean')

# Display the plot
plt.show()
```

```python
# Determine STD corresponding to highest mean
highest_mean_index = df['Sim Mean'].idxmax()
highest_mean_std = df.loc[highest_mean_index, 'Sim STD']

print(highest_mean_std)

# Separate into two DFs, those with standard deviation below the standard deviation of the highest return and those with standard deviation above
df_low = df[df['Sim STD'] <= highest_mean_std]
df_high = df[df['Sim STD'] > highest_mean_std]

df_low.reset_index(inplace=True)
df_high.reset_index(inplace=True)


print(len(df_low), len(df_high))
```

```python
# Check for Pareto optimality - Maximum Return, Minimum Standard Deviation
is_pareto_optimal = []
for i in range(len(df_low)):
    current_mean = df_low.loc[i, 'Sim Mean']
    current_std = df_low.loc[i, 'Sim STD']
    is_pareto = True
    for j in range(len(df_low)):
        if i != j:
            if df_low.loc[j, 'Sim Mean'] >= current_mean and df_low.loc[j, 'Sim STD'] <= current_std:
                is_pareto = False
                break
    is_pareto_optimal.append(is_pareto)

# Filter out non-Pareto optimal rows
df_pareto_optimal_low = df_low[is_pareto_optimal]

# Print the resulting DataFrame
df_pareto_optimal_low.sort_values(by=['Sim STD'], inplace=True)
```

```python
# Check for Pareto optimality - Maximum Return, Maximum Standard Deviation
is_pareto_optimal = []
for i in range(len(df_high)):
    current_mean = df_high.loc[i, 'Sim Mean']
    current_std = df_high.loc[i, 'Sim STD']
    is_pareto = True
    for j in range(len(df_high)):
        if i != j:
            if df_high.loc[j, 'Sim Mean'] >= current_mean and df_high.loc[j, 'Sim STD'] >= current_std:
                is_pareto = False
                break
    is_pareto_optimal.append(is_pareto)

# Filter out non-Pareto optimal rows
df_pareto_optimal_high = df_high[is_pareto_optimal]

# Print the resulting DataFrame
df_pareto_optimal_high.sort_values(by=['Sim STD'], inplace=True)
```

```python
df_pareto_optimal = pd.concat([df_pareto_optimal_low, df_pareto_optimal_high], axis=1)

# Assuming your DataFrame is named 'df'
x = df_pareto_optimal['Sim STD']
y = df_pareto_optimal['Sim Mean']

# Create a scatter plot
plt.scatter(x, y)

for j, txt in enumerate(x):
    plt.annotate(txt, (x[j], y[j]))

# Set labels and title
plt.xlabel('Sim STD')
plt.ylabel('Sim Mean')
plt.title('Scatter Plot: Sim STD vs. Sim Mean')

# Display the plot
plt.show()
```
