```python
# Read in actual game scores
def create_scores(team_map, start_date, end_date):
    team_map = team_map[['FULLNAME', 'BBREFTEAM']]
    
    games = statsapi.schedule(start_date=start_date, end_date=end_date)
    
    all_games = []
    for game in games:
        game_list = [game['game_id'], game['game_type'], game['away_name'], game['home_name'], game['away_score'], game['home_score'], game['game_date']]
        all_games.append(game_list)

    all_games_df = pd.DataFrame(all_games, columns=['game_id', 'game_type', 'away_name', 'home_name', 'away_score', 'home_score', 'game_date'])
    
    all_games_df = all_games_df.query('game_type == "R"')
    
    all_games_df = all_games_df.merge(team_map, left_on='away_name', right_on='FULLNAME', how='left')
    all_games_df = all_games_df.merge(team_map, left_on='home_name', right_on='FULLNAME', how='left')
    
    all_games_df.rename(columns={'BBREFTEAM_x':'away_team', 'BBREFTEAM_y':'home_team','game_date':'date'},inplace=True)
    all_games_df.drop(columns={'FULLNAME_x', 'FULLNAME_y', 'away_name', 'home_name'},inplace=True)
    
    all_games_df['date'] = all_games_df['date'].str.replace("-", "")

    
    return all_games_df
```

```python
def create_scores2(team_map, start_date=todaysdate_slash, end_date=todaysdate_slash):
    start_year = start_date[-4:]
    end_year = end_date[-4:]
    
    if start_year != end_year:
        scores2022 = create_scores(team_map, "03/30/2022", "10/31/2022")
        scores2023 = create_scores(team_map, "03/30/2023", end_date)
        scores = pd.concat([scores2022, scores2023], axis=0)
        scores.reset_index(drop=True, inplace=True)
    else:
        scores = create_scores(team_map, start_date, end_date)
        
    # Number games (in case of double headers)
    scores['game'] = scores.groupby(['away_team', 'home_team', 'date']).cumcount() + 1

    
    return scores
```

```python
def read_gamble(start_date=todaysdate, end_date=todaysdate):
    folder_path = r'C:\Users\james\Documents\MLB\Data2\11. Projections\C. Odds'
    file_prefix = 'Odds'
    date_format = '%Y%m%d'
    
    # Convert input dates to datetime objects
    start_datetime = pd.to_datetime(start_date, format=date_format)
    end_datetime = pd.to_datetime(end_date, format=date_format)
    
    # Initialize an empty list to store DataFrames from each file
    dfs = []
    
    # Loop through the files in the folder
    for filename in os.listdir(folder_path):
        # Check if the file matches the pattern "Odds _______.csv"
        if filename.startswith(file_prefix) and filename.endswith('.csv'):
            file_date_str = filename[len(file_prefix)+1:-4]  # Extract the date part
            file_datetime = pd.to_datetime(file_date_str, format=date_format)
            
            # Check if the file's date is within the specified range
            if start_datetime <= file_datetime <= end_datetime:
                file_path = os.path.join(folder_path, filename)
                df = pd.read_csv(file_path)
                dfs.append(df)
    
    # Concatenate all DataFrames into a single DataFrame
    if dfs:
        result_df = pd.concat(dfs, axis=0, ignore_index=True)

        # Number games (in case of double headerS)
        result_df = result_df.iloc[::-1]
        result_df['game'] = result_df.groupby(['date', 'away_team', 'home_team']).cumcount() + 1
        result_df = result_df[['away_team', 'away_runs', 'home_team', 'home_runs', 'ou',
       'ou_juice', 'spread', 'spread_juice', 'away_ml', 'home_ml', 'date',
       'game']]
        
        result_df['date'] = result_df['date'].astype('str')
        return result_df
    else:
        print("No files found within the specified date range.")
        return None
```

```python
def calculate_percentages(file_path, file_name):
    
    df = pd.read_excel(os.path.join(file_path, file_name))
    
    df['AwayWin%'] = (df['Away'] > df['Home']).astype(int)
    df['AwayWin0.0%'] = ((df['Away'] - df['Home']) > 0).astype(int)
    
    df['AwayWin0.5%'] = ((df['Away'] - df['Home']) > 0.5).astype(int)
    df['AwayWin1.0%'] = ((df['Away'] - df['Home']) > 1).astype(int)
    df['AwayWin1.5%'] = ((df['Away'] - df['Home']) > 1.5).astype(int)
    df['AwayWin2.0%'] = ((df['Away'] - df['Home']) > 2).astype(int)
    
    df['AwayWin-0.5%'] = ((df['Away'] - df['Home']) > -0.5).astype(int)
    df['AwayWin-1.0%'] = ((df['Away'] - df['Home']) > -1).astype(int)
    df['AwayWin-1.5%'] = ((df['Away'] - df['Home']) > -1.5).astype(int)
    df['AwayWin-2.0%'] = ((df['Away'] - df['Home']) > -2).astype(int)

    # Calculate the percentages for every value from 5 to 15, going by half-point increments
    for value in range(5, 16):
        value = value + 0.0
        column_name = f'Over{value}'
        df[column_name] = ((df['Away'] + df['Home']) > value).mean()
        
        column_name = f'Over{value + 0.5}'
        df[column_name] = ((df['Away'] + df['Home']) > value + 0.5).mean()

        column_name = f'Under{value}'
        df[column_name] = ((df['Away'] + df['Home']) < value).mean()

        column_name = f'Under{value + 0.5}'
        df[column_name] = ((df['Away'] + df['Home']) < value + 0.5).mean()

    # Calculate the averages of all columns (percentages and additional metrics)
    averages = df.mean()

    # Create a new DataFrame with one row containing the averages and additional information
    result_df = pd.DataFrame(averages).transpose()
    
    result_df['file'] = file_name
    result_df['away_team'] = file_name[0:3]
    result_df['home_team'] = file_name[4:7]
    result_df['date'] = pd.to_datetime(file_name[8:16], format='%m%d%Y').strftime('%Y%m%d')
    result_df['time'] = file_name[17:23]
    
    
    return result_df

# Maybe make this take dates ranges?
def read_projections(start_date, end_date):
    
    history = pd.read_csv(os.path.join(baseball_path, "Utilities", "Contests.csv"), dtype='str')
    history = history[['contestKey', 'date']]
    
    # Filter the DataFrame based on the given date range
    filtered_contestKeys = history.loc[(history['date'] >= start_date) & (history['date'] <= end_date), 'contestKey'].tolist()
    
    
    # Define the main directory path
    main_dir = r'C:\Users\james\Documents\MLB\Data2\10. Sims\B. Scores'

    # Initialize an empty list to store all result DataFrames
    all_result_dfs = []
    
    for contestKey in filtered_contestKeys:
        folder = "Matchups" + contestKey
        
        file_path = os.path.join(main_dir, folder)
        
        if os.path.isdir(file_path):
            for file_name in os.listdir(file_path):
                    # Call the calculate_percentages function to get the result DataFrame for the current file
                    result_df = calculate_percentages(file_path, file_name)

                    # Append the result DataFrame to the list
                    all_result_dfs.append(result_df)
                
    # Concatenate all result DataFrames into one final DataFrame
    final_df = pd.concat(all_result_dfs, ignore_index=True)

    final_df.drop_duplicates('file', inplace=True)
    final_df.reset_index(inplace=True, drop=True)

    final_df.drop(columns={'Unnamed: 0'}, inplace=True)

    return final_df           
```

```python
def gambling_dashboard(team_map, start_date, end_date):
    
    start_date_slash = start_date[4:6] + "/" + start_date[6:9] + "/" + start_date[0:4]
    end_date_slash = end_date[4:6] + "/" + end_date[6:9] + "/" + end_date[0:4]
    
    # Gambling odds
    odds = read_gamble(start_date, end_date)
    projected = read_projections(start_date, end_date)
    scores = create_scores2(team_map, start_date_slash, end_date_slash)

    merged_df = pd.merge(odds, scores, on=['away_team', 'home_team', 'date', 'game'], how='inner')
    merged_df = pd.merge(projected, merged_df, on=['away_team', 'home_team', 'date'], how='inner')
    
    merged_df.dropna(inplace=True)
    

    # Create a new column 'OverWin%' using apply() and lambda function
    merged_df['OverWin%'] = merged_df.apply(lambda row: row[f'Over{row["ou"]}'], axis=1)
    merged_df['UnderWin%'] = merged_df.apply(lambda row: row[f'Under{row["ou"]}'], axis=1)
    merged_df['AwayCover%'] = merged_df.apply(lambda row: row[f'AwayWin{row["spread"]}%'], axis=1)

    merged_df['HomeWin%'] = 1 - merged_df['AwayWin%']
    merged_df['HomeCover%'] = 1 - merged_df['AwayCover%']

    merged_df['AwayWin$'] = (merged_df.apply(lambda row: (100 / abs(row['away_ml'])) * 100 if row['away_ml'] < 0 else (row['away_ml'] / 100) * 100, axis=1) + 100)
    merged_df['HomeWin$'] = (merged_df.apply(lambda row: (100 / abs(row['home_ml'])) * 100 if row['home_ml'] < 0 else (row['home_ml'] / 100) * 100, axis=1) + 100)
    
    merged_df['AwayWin$Exp'] = merged_df['AwayWin$'] * merged_df['AwayWin%'] 
    merged_df['HomeWin$Exp'] = merged_df['HomeWin$'] * merged_df['HomeWin%'] 
    
    merged_df['AwayWin$Act'] = merged_df['AwayWin$'] * (merged_df['away_score'] > merged_df['home_score'])
    merged_df['HomeWin$Act'] = merged_df['HomeWin$'] * (merged_df['away_score'] < merged_df['home_score']) 
    

    merged_df = merged_df[['date', 'away_team', 'home_team', 'Away', 'Home', 'away_score', 'home_score', 'AwayWin%', 'HomeWin%', 'AwayWin$Exp', 'HomeWin$Exp', 'AwayWin$Act', 'HomeWin$Act']]

    return merged_df
```
