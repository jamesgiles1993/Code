```python
def create_game(matchup, matchup_path, team_map, weather_df):
    AwayBatters, HomePitchers, HomeBatters, AwayPitchers, park = create_matchup(matchup, matchup_path)
    venue = team_map.loc[park, 'VENUE_ID']
    game = Scoreboard(AwayBatters, HomePitchers, HomeBatters, AwayPitchers, park, 9)

    venue_id = "venue_" + str(venue)
    
    # Loop over all the variables in the game object
    for attr, value in vars(game).items():
        # If the attribute is the venue ID
        if attr == venue_id:
            # Set it to 1
            value = 1
            setattr(game, attr, value)
        
    game.TEMP_PARK_CT = weather_df.loc[venue]['TEMP_PARK_CT']
    game.x_vect = weather_df.loc[venue]['x_vect']
    game.y_vect = weather_df.loc[venue]['y_vect']
    
    return game
```

```python
# Create PA event probability
def probabilities(game, model_binary, model_outs, model_safe):
    # If the pitcher is a lefty, use batter's vs L stats
    if game.pitching.throws == "L":
        # Short
        game.ab.b1 = game.ab.b1_l
        game.ab.b2 = game.ab.b2_l
        game.ab.b3 = game.ab.b3_l
        game.ab.hr = game.ab.hr_l
        game.ab.bb = game.ab.bb_l
        game.ab.hbp = game.ab.hbp_l
        
        game.ab.so = game.ab.so_l
        game.ab.lo = game.ab.lo_l
        game.ab.po = game.ab.po_l
        game.ab.go = game.ab.go_l
        game.ab.fo = game.ab.fo_l
        
        game.ab.woba = game.ab.woba_l
        game.ab.slg = game.ab.slg_l
        game.ab.obp = game.ab.obp_l
        game.ab.iso = game.ab.iso_l
        
        game.ab.to_left = game.ab.to_left_l
        game.ab.to_middle = game.ab.to_middle_l
        game.ab.to_right = game.ab.to_right_l
        
        game.ab.hard_hit = game.ab.hard_hit_l
        game.ab.totalDistance = game.ab.totalDistance_l
        game.ab.launchSpeed = game.ab.launchSpeed_l
        
        game.ab.pa = game.ab.pa_l
        game.ab.ab = game.ab.ab_l
        
        # Long
        game.ab.b1_long = game.ab.b1_long_l
        game.ab.b2_long = game.ab.b2_long_l
        game.ab.b3_long = game.ab.b3_long_l
        game.ab.hr_long = game.ab.hr_long_l
        game.ab.bb_long = game.ab.bb_long_l
        game.ab.hbp_long = game.ab.hbp_long_l
        
        game.ab.so_long = game.ab.so_long_l
        game.ab.lo_long = game.ab.lo_long_l
        game.ab.po_long = game.ab.po_long_l
        game.ab.go_long = game.ab.go_long_l
        game.ab.fo_long = game.ab.fo_long_l
        
        game.ab.woba_long = game.ab.woba_long_l
        game.ab.slg_long = game.ab.slg_long_l
        game.ab.obp_long = game.ab.obp_long_l
        game.ab.iso_long = game.ab.iso_long_l
        
        game.ab.to_left_long = game.ab.to_left_long_l
        game.ab.to_middle_long = game.ab.to_middle_long_l
        game.ab.to_right_long = game.ab.to_right_long_l
        
        game.ab.hard_hit_long = game.ab.hard_hit_long_l
        game.ab.totalDistance_long = game.ab.totalDistance_long_l
        game.ab.launchSpeed_long = game.ab.launchSpeed_long_l
        
        game.ab.pa_long = game.ab.pa_long_l
        game.ab.ab_long = game.ab.ab_long_l
        
    # If not, use stats vs R
    else:
        # Short
        game.ab.b1 = game.ab.b1_r
        game.ab.b2 = game.ab.b2_r
        game.ab.b3 = game.ab.b3_r
        game.ab.hr = game.ab.hr_r
        game.ab.bb = game.ab.bb_r
        game.ab.hbp = game.ab.hbp_r
        
        game.ab.so = game.ab.so_r
        game.ab.lo = game.ab.lo_r
        game.ab.po = game.ab.po_r
        game.ab.go = game.ab.go_r
        game.ab.fo = game.ab.fo_r
        
        game.ab.woba = game.ab.woba_r
        game.ab.slg = game.ab.slg_r
        game.ab.obp = game.ab.obp_r
        game.ab.iso = game.ab.iso_r
        
        game.ab.to_left = game.ab.to_left_r
        game.ab.to_middle = game.ab.to_middle_r
        game.ab.to_right = game.ab.to_right_r
        
        game.ab.hard_hit = game.ab.hard_hit_r
        game.ab.totalDistance = game.ab.totalDistance_r
        game.ab.launchSpeed = game.ab.launchSpeed_r
        
        game.ab.pa = game.ab.pa_r
        game.ab.ab = game.ab.ab_r
        
        # Long
        game.ab.b1_long = game.ab.b1_long_r
        game.ab.b2_long = game.ab.b2_long_r
        game.ab.b3_long = game.ab.b3_long_r
        game.ab.hr_long = game.ab.hr_long_r
        game.ab.bb_long = game.ab.bb_long_r
        game.ab.hbp_long = game.ab.hbp_long_r
        
        game.ab.so_long = game.ab.so_long_r
        game.ab.lo_long = game.ab.lo_long_r
        game.ab.po_long = game.ab.po_long_r
        game.ab.go_long = game.ab.go_long_r
        game.ab.fo_long = game.ab.fo_long_r
        
        game.ab.woba_long = game.ab.woba_long_r
        game.ab.slg_long = game.ab.slg_long_r
        game.ab.obp_long = game.ab.obp_long_r
        game.ab.iso_long = game.ab.iso_long_r
        
        game.ab.to_left_long = game.ab.to_left_long_r
        game.ab.to_middle_long = game.ab.to_middle_long_r
        game.ab.to_right_long = game.ab.to_right_long_r
        
        game.ab.hard_hit_long = game.ab.hard_hit_long_r
        game.ab.totalDistance_long = game.ab.totalDistance_long_r
        game.ab.launchSpeed_long = game.ab.launchSpeed_long_r
        
        game.ab.pa_long = game.ab.pa_long_r
        game.ab.ab_long = game.ab.ab_long_r
        
    # # This is pointless lol - don't need to set things with defaults, sb, fg - just make sure 
    # game.ab.sba_2b = game.ab.sba_2b
    # game.ab.sb_2b = game.ab.sb_2b
    # game.ab.sba_3b = game.ab.sba_3b
    # game.ab.sb_3b = game.ab.sb_3b

    # Determine hand for switch hitters
    if game.ab.bats_copy == "S":
        if game.pitching.throws == "L":
            game.ab.bats = "R"
        else:
            game.ab.bats = "L"

    # Pitcher Stats            
    # If the batter is a lefty, use the pitcher's vs L stats
    if game.ab.bats == "L":
        # Short
        game.pitching.b1 = game.pitching.b1_l
        game.pitching.b2 = game.pitching.b2_l
        game.pitching.b3 = game.pitching.b3_l
        game.pitching.hr = game.pitching.hr_l
        game.pitching.bb = game.pitching.bb_l
        game.pitching.hbp = game.pitching.hbp_l
        
        game.pitching.so = game.pitching.so_l
        game.pitching.lo = game.pitching.lo_l
        game.pitching.po = game.pitching.po_l
        game.pitching.go = game.pitching.go_l
        game.pitching.fo = game.pitching.fo_l
        
        game.pitching.woba = game.pitching.woba_l
        game.pitching.slg = game.pitching.slg_l
        game.pitching.obp = game.pitching.obp_l
        game.pitching.iso = game.pitching.iso_l
        
        game.pitching.to_left = game.pitching.to_left_l
        game.pitching.to_middle = game.pitching.to_middle_l
        game.pitching.to_right = game.pitching.to_right_l
        
        game.pitching.hard_hit = game.pitching.hard_hit_l
        game.pitching.maxSpeed = game.pitching.maxSpeed_l
        game.pitching.maxSpin = game.pitching.maxSpin_l
        
        game.pitching.pa = game.pitching.pa_l
        game.pitching.ab = game.pitching.ab_l
        
        # Long
        game.pitching.b1_long = game.pitching.b1_long_l
        game.pitching.b2_long = game.pitching.b2_long_l
        game.pitching.b3_long = game.pitching.b3_long_l
        game.pitching.hr_long = game.pitching.hr_long_l
        game.pitching.bb_long = game.pitching.bb_long_l
        game.pitching.hbp_long = game.pitching.hbp_long_l
        
        game.pitching.so_long = game.pitching.so_long_l
        game.pitching.lo_long = game.pitching.lo_long_l
        game.pitching.po_long = game.pitching.po_long_l
        game.pitching.go_long = game.pitching.go_long_l
        game.pitching.fo_long = game.pitching.fo_long_l
        
        game.pitching.woba_long = game.pitching.woba_long_l
        game.pitching.slg_long = game.pitching.slg_long_l
        game.pitching.obp_long = game.pitching.obp_long_l
        game.pitching.iso_long = game.pitching.iso_long_l
        
        game.pitching.to_left_long = game.pitching.to_left_long_l
        game.pitching.to_middle_long = game.pitching.to_middle_long_l
        game.pitching.to_right_long = game.pitching.to_right_long_l
        
        game.pitching.hard_hit_long = game.pitching.hard_hit_long_l
        game.pitching.maxSpeed_long = game.pitching.maxSpeed_long_l
        game.pitching.maxSpin_long = game.pitching.maxSpin_long_l
        
        game.pitching.pa_long = game.pitching.pa_long_l
        game.pitching.ab_long = game.pitching.ab_long_l
        
    # If not, use stats vs R
    else:
        # Short
        game.pitching.b1 = game.pitching.b1_r
        game.pitching.b2 = game.pitching.b2_r
        game.pitching.b3 = game.pitching.b3_r
        game.pitching.hr = game.pitching.hr_r
        game.pitching.bb = game.pitching.bb_r
        game.pitching.hbp = game.pitching.hbp_r
        
        game.pitching.so = game.pitching.so_r
        game.pitching.lo = game.pitching.lo_r
        game.pitching.po = game.pitching.po_r
        game.pitching.go = game.pitching.go_r
        game.pitching.fo = game.pitching.fo_r
        
        game.pitching.woba = game.pitching.woba_r
        game.pitching.slg = game.pitching.slg_r
        game.pitching.obp = game.pitching.obp_r
        game.pitching.iso = game.pitching.iso_r
        
        game.pitching.to_left = game.pitching.to_left_r
        game.pitching.to_middle = game.pitching.to_middle_r
        game.pitching.to_right = game.pitching.to_right_r
        
        game.pitching.hard_hit = game.pitching.hard_hit_r
        game.pitching.maxSpeed = game.pitching.maxSpeed_r
        game.pitching.maxSpin = game.pitching.maxSpin_r
        
        game.pitching.pa = game.pitching.pa_r
        game.pitching.ab = game.pitching.ab_r
        
        # Long
        game.pitching.b1_long = game.pitching.b1_long_r
        game.pitching.b2_long = game.pitching.b2_long_r
        game.pitching.b3_long = game.pitching.b3_long_r
        game.pitching.hr_long = game.pitching.hr_long_r
        game.pitching.bb_long = game.pitching.bb_long_r
        game.pitching.hbp_long = game.pitching.hbp_long_r
        
        game.pitching.so_long = game.pitching.so_long_r
        game.pitching.lo_long = game.pitching.lo_long_r
        game.pitching.po_long = game.pitching.po_long_r
        game.pitching.go_long = game.pitching.go_long_r
        game.pitching.fo_long = game.pitching.fo_long_r
        
        game.pitching.woba_long = game.pitching.woba_long_r
        game.pitching.slg_long = game.pitching.slg_long_r
        game.pitching.obp_long = game.pitching.obp_long_r
        game.pitching.iso_long = game.pitching.iso_long_r
        
        game.pitching.to_left_long = game.pitching.to_left_long_r
        game.pitching.to_middle_long = game.pitching.to_middle_long_r
        game.pitching.to_right_long = game.pitching.to_right_long_r
        
        game.pitching.hard_hit_long = game.pitching.hard_hit_long_r
        game.pitching.maxSpeed_long = game.pitching.maxSpeed_long_r
        game.pitching.maxSpin_long = game.pitching.maxSpin_long_r
        
        game.pitching.pa_long = game.pitching.pa_long_r
        game.pitching.ab_long = game.pitching.ab_long_r
    
    if game.ab.bats == "L":
        batter_hand = 'L_Position'
    else: 
        batter_hand = 'R_Position'
    if game.pitching.throws == "L":
        pitcher_hand = "_l"
    else:
        pitcher_hand = "_r"
    
    # Make compatible with model
    if game.ab.bats == "L":
        game.ab.b_L = 1
    else:
        game.ab.b_L = 0
    
    if game.pitching.throws == "L":
        game.pitching.p_L = 1
    else:
        game.pitching.p_L = 0
        
    # On base dummies
    if game.on_1b is not None:
        game.onFirst = 1
    else:
        game.onFirst = 0
    if game.on_2b is not None:
        game.onSecond = 1
    else:
        game.onSecond = 0
    if game.on_3b is not None:
        game.onThird = 1
    else: 
        game.onThird = 0
        
    if game.top_bot == "Top":
        game.top = 1
    else:
        game.top = 0

    if game.top == 1:
        game.score_diff = game.away_score - game.home_score
    else:
        game.score_diff = game.home_score - game.away_score
    
    game.year_2013 = 0
    game.year_2014 = 0
    game.year_2015 = 0
    game.year_2016 = 0
    game.year_2017 = 0
    game.year_2018 = 0
    game.year_2019 = 0
    game.year_2020 = 0
    game.year_2021 = 0
    game.year_2022 = 0
    game.year_2023 = 0

    
    
    # Inputs
    batter_stats = [game.ab.b1, game.ab.b2, game.ab.b3, game.ab.hr, game.ab.bb, game.ab.hbp, 
                    game.ab.so, game.ab.fo, game.ab.go, game.ab.lo, game.ab.po,
                    game.ab.woba, game.ab.slg, game.ab.obp, game.ab.iso, 
                    game.ab.to_left, game.ab.to_middle, game.ab.to_right,
                    game.ab.hard_hit, game.ab.totalDistance, game.ab.launchSpeed, 
                    
                    game.ab.b1_long, game.ab.b2_long, game.ab.b3_long, game.ab.hr_long, game.ab.bb_long, game.ab.hbp_long,
                    game.ab.so_long, game.ab.fo_long, game.ab.go_long, game.ab.lo_long, game.ab.po_long,
                    game.ab.woba_long, game.ab.slg_long, game.ab.obp_long, game.ab.iso_long, 
                    game.ab.to_left_long, game.ab.to_middle_long, game.ab.to_right_long,
                    game.ab.hard_hit_long, game.ab.totalDistance_long, game.ab.launchSpeed_long]

    # Removed game.ab.ab, game.ab.pa, game.ab.ab_long, game.ab.pa_long
    
    pitcher_stats = [game.pitching.b1, game.pitching.b2, game.pitching.b3, game.pitching.hr, game.pitching.bb, game.pitching.hbp, 
                    game.pitching.so, game.pitching.fo, game.pitching.go, game.pitching.lo, game.pitching.po,
                    game.pitching.woba, game.pitching.slg, game.pitching.obp, game.pitching.iso, 
                    game.pitching.to_left, game.pitching.to_middle, game.pitching.to_right,
                    game.pitching.hard_hit, game.pitching.maxSpeed, game.pitching.maxSpin, 
                     
                    game.pitching.b1_long, game.pitching.b2_long, game.pitching.b3_long, game.pitching.hr_long, game.pitching.bb_long, game.pitching.hbp_long,
                    game.pitching.so_long, game.pitching.fo_long, game.pitching.go_long, game.pitching.lo_long, game.pitching.po_long,
                    game.pitching.woba_long, game.pitching.slg_long, game.pitching.obp_long, game.pitching.iso_long, 
                    game.pitching.to_left_long, game.pitching.to_middle_long, game.pitching.to_right_long,
                    game.pitching.hard_hit_long, game.pitching.maxSpeed_long, game.pitching.maxSpin_long]
    
    # Removed game.pitching.ab, game.pitching.pa, game.pitching.ab_long, game.pitching.pa_long

    batter_stats_fg = [game.ab.b1_rate, game.ab.b2_rate, game.ab.b3_rate, game.ab.hr_rate, game.ab.bb_rate, game.ab.hbp_rate, game.ab.so_rate,
                       game.ab.fgwoba, game.ab.fgslg, game.ab.fgobp]

    pitcher_stats_fg = [game.pitching.H9, game.pitching.HR9, game.pitching.K9, game.pitching.BB9]


    venues = [game.venue_1, game.venue_2, game.venue_3, game.venue_4, game.venue_5, game.venue_7, game.venue_10, game.venue_12, 
              game.venue_13, game.venue_14, game.venue_15, game.venue_16, game.venue_17, game.venue_19, game.venue_22, game.venue_31,
              game.venue_32, game.venue_680, game.venue_2392, game.venue_2394, game.venue_2395, game.venue_2535, game.venue_2536,
              game.venue_2602, game.venue_2680, game.venue_2681, game.venue_2701, game.venue_2735, game.venue_2756, game.venue_2889, 
              game.venue_3289, game.venue_3309, game.venue_3312, game.venue_3313, game.venue_4169, game.venue_4705, game.venue_5010, 
              game.venue_5325, game.venue_5365, game.venue_5381, game.venue_5445]

    years = [game.year_2015, game.year_2016, game.year_2017, game.year_2018, game.year_2019, game.year_2020, game.year_2021, game.year_2022, game.year_2023]

    other_list = [game.pitching.p_L, game.ab.b_L, 
                  game.x_vect, game.y_vect, game.TEMP_PARK_CT, game.onFirst, game.onSecond, game.onThird, game.inning, game.top, game.score_diff]
    
    ### Create inputs 
    inputs = batter_stats + pitcher_stats + venues + years + other_list    
    model_inputs = pd.Series(inputs).values.reshape(1,-1)
    

    
    # Separate outs and safe
    binary_list = model_binary.predict_proba(model_inputs).tolist()
    outs_list = model_outs.predict_proba(model_inputs).tolist()
    safe_list = model_safe.predict_proba(model_inputs).tolist()
    
    is_safe = binary_list[0][0]
    is_out = 1 - is_safe
    
    fo_pred = outs_list[0][0]
    go_pred = outs_list[0][1]
    lo_pred = outs_list[0][2]
    po_pred = outs_list[0][3]
    so_pred = outs_list[0][4]
    
    b1_pred = safe_list[0][0]
    b2_pred = safe_list[0][1]
    b3_pred = safe_list[0][2]
    bb_pred = safe_list[0][3]
    hbp_pred = safe_list[0][4]
    hr_pred = safe_list[0][5]
       
#     # One model
#     binary_list = model_binary.predict_proba(model_inputs).tolist()
#     # outs_list = model_outs.predict_proba(model_inputs).tolist()
#     # safe_list = model_safe.predict_proba(model_inputs).tolist()
#     is_safe = 1
#     is_out = 1
    
    
#     b1_pred = binary_list[0][0]
#     b2_pred = binary_list[0][1]
#     b3_pred = binary_list[0][2]
#     bb_pred = binary_list[0][3]
#     fo_pred = binary_list[0][4]
#     go_pred = binary_list[0][5]
#     hbp_pred = binary_list[0][6]
#     hr_pred = binary_list[0][7]
#     lo_pred = binary_list[0][8]
#     po_pred = binary_list[0][9]
#     so_pred = binary_list[0][10]
    
    b1 = is_safe * b1_pred
    b2 = is_safe * b2_pred
    b3 = is_safe * b3_pred
    bb = is_safe * bb_pred
    fo = is_out * fo_pred
    go = is_out * go_pred
    hbp = is_safe * hbp_pred
    hr = is_safe * hr_pred
    lo = is_out * lo_pred
    po = is_out * po_pred
    so = is_out * so_pred 
    
    
       
    # Add stats together so they represent the end point of a range with a probability of being selected equal to their rate
    bb = hbp + bb 
    b1 = bb + b1
    b2 = b1 + b2
    b3 = b2 + b3
    hr = b3 + hr
    so = hr + so
    lo = so + lo
    po = lo + po
    go = po + go
    fo = go + fo
    # Set flyout to be 1. This model almost guarantees this. It's accurate to like, 10 decimal places, but why risk it?
    fo = 1
    
    del model_inputs
    
    return hbp, bb, b1, b2, b3, hr, so, lo, po, go, fo, game 
```
