```python
import numpy as np
import pandas as pd
import os

from datetime import datetime

baseball_path = r"C:\Users\james\Documents\MLB\Data"
```

```python
history = pd.read_csv(os.path.join(baseball_path, "Y. One Time Use", "draftkings-contest-entry-history.csv"))                  
```

```python
df = history.query('Entry_Fee == "$4.00 "').query('Sport == "MLB"')
df['date'] = pd.to_datetime(df['Contest_Date_EST'], format='%m/%d/%Y %H:%M').dt.strftime('%Y%m%d')

df[['Contest_Date_EST', 'Contest_Key', 'Contest_Entries', 'date']]
```

```python
subset = df[df['date'] >= "20230101"]
contestKeys = list(subset['Contest_Key'])
contestKeys = [i for n, i in enumerate(contestKeys) if i not in contestKeys[:n]]

contestKeys
```

```python
df['Match'] = 0  # Initialize the "Match" column with 0

for index, row in df.iterrows():
    date = row['Contest_Date_EST']
    datetime_obj = datetime.strptime(date, "%m/%d/%Y %H:%M")
    formatted_date = datetime_obj.strftime("%Y%m%d")
    filename = f"Results {formatted_date}.csv"

    try:
        # Read the CSV file
        result_df = pd.read_csv(os.path.join(baseball_path, "Results", filename))
        result_length = len(result_df)

        # Compare the result length with Contest_Entries
        contest_entries = row['Contest_Entries']
        if abs(result_length - contest_entries) <= 5:
            df.at[index, 'Match'] = 1  # Set the "Match" column value to 1

        # Optionally, you can print the match status
        print(f"Match for {formatted_date}: {df.at[index, 'Match']}")
        
    except FileNotFoundError:
        print(f"CSV file not found for {formatted_date}")
```
